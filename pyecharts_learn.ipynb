```python
#coding:utf8
import pyecharts
import pandas as pd
from pyecharts.charts import Bar
from pyecharts.charts import Pie
from pyecharts import options as opts
```

```python
f=open('Task_4_lianjia_sale.csv')
lianjia_data = pd.read_csv(f)
lianjia_data
df_avg = lianjia_data.groupby('区域')['价格'].agg([('房源数量','count'),('区域平均价格','mean'),('最高价','max'),('最低价','min')]).round(0)
df_avg
```

```python
df_avg=df_avg.sort_values(by=['区域平均价格'])
df_avg.head()
```

```python
from pyecharts import options as opts
bar=Bar()
xaxis=df_avg.index.tolist()
yaxis=df_avg['区域平均价格'].tolist()
bar.add_xaxis(xaxis)
bar.add_yaxis('区域平均价格',yaxis)
bar.set_global_opts(title_opts=opts.TitleOpts(
title="上海各区域平均价格" #设置主标题
# subtitle='', #设置副标题
))
bar.render_notebook()
```
