```python
# coding:utf8
import pandas as pd
import tushare as ts
from pyecharts import options as opts
```

```python
df = ts.get_stock_basics()
df
```

```python
f=open('Task_4_lianjia_sale.csv')
lianjia_data = pd.read_csv(f)
f.close()
df_avg = lianjia_data.groupby('区域')['价格'].agg([('区域平均价格','mean')]).round(0)
price = df_avg.values.tolist()
districts = df_avg.index.tolist()
districts_new = [i+'区' if i!='浦东' else i+'新区' for i in df_avg.index.tolist()]  #格式化地区名称
data=[list(i) for i in zip(districts_new,price)]

map_sh = Map()
map_sh.add("", data, maptype="上海")
map_sh.set_global_opts(
    title_opts=opts.TitleOpts(title="上海住房平均价格表"),
    visualmap_opts=opts.VisualMapOpts(max_=100000,min_=18000)  #数据范围
    )
map_sh.set_series_opts(label_opts=opts.LabelOpts(is_show=False))  #不显示标签
map_sh.render_notebook()
```

```python
data_company = df.groupby('area')['area'].agg([('上市公司数量','count')])
province_list = [i+'古' if i =='内蒙' else i for i in data_company.index.tolist()]
company_count = data_company['上市公司数量'].tolist()
data_count = [list(i) for i in zip(province_list,company_count)]
data_count
map_company = Map()
map_company.add('上市公司数量',data_count, maptype='china')
map_company.set_global_opts(
    title_opts=opts.TitleOpts(title="中国各省上市公司数量"),
    visualmap_opts=opts.VisualMapOpts(max_=500))
map_company.render_notebook()
```
