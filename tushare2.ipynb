```python
# coding:utf8
import pandas as pd
import tushare as ts
from pyecharts import options as opts
```

```python
df = ts.get_stock_basics()
df
```

```python
gpr_mean = df['gpr'].mean()
npr_mean = df['npr'].mean()
```

```python
from pyecharts.charts import Liquid
lq_pr = Liquid()
lq_pr.add("", [gpr_mean/100,gpr_mean/100], shape='rect', is_outline_show=True)
        # 水球外形，有' circle', 'rect', 'roundRect', 'triangle', 'diamond', 'pin', 'arrow' 可选。
        # 默认 'circle'。也可以为自定义的 SVG 路径。
        #is_outline_show设置边框
lq_pr.set_global_opts(title_opts=opts.TitleOpts(title="毛利率 水球图"))
lq_pr.render_notebook()
```

```python
from pyecharts.charts import Gauge
gauge_pr = Gauge()
gauge_pr.add('',[['净利率',round(npr_mean,2)]],max_=40,min_=-40)
lq_pr.set_global_opts(title_opts=opts.TitleOpts(title="净利率 仪表盘"))
gauge_pr.render_notebook()
```

```python
df_industry = df.groupby('industry')['name'].agg([('数量','count')])
data_industry = [list(i) for i in zip(df_industry.index.tolist(),df_industry['数量'].tolist())]
from pyecharts.charts import WordCloud
wordcloud_industry = WordCloud()
wordcloud_industry.add('',data_industry) #shape='diamond' 可设置参数shape来控制形状
wordcloud_industry.set_global_opts(title_opts=opts.TitleOpts(title="中国上市公司行业词云图"))
wordcloud_industry.render_notebook()
```

```python
df_600519 = ts.get_hist_data('600519',start='',end='20200415')
df_600519.sort_index(inplace=True)
```

```python
data_high = df_600519['high'].tolist()
data_low = df_600519['low'].tolist()
data_index = df_600519.index.tolist()
from pyecharts.charts import Line
price_line = Line()
price_line.add_xaxis(data_index)
#label_opts=opts.LabelOpts(is_show=False) 控制是否在线上显示各个点的值
#markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="min")]) 标注线用来追踪屏幕内最大最小值
price_line.add_yaxis('最低价',data_low,is_smooth=True,label_opts=opts.LabelOpts(is_show=False),markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="min")]))
price_line.add_yaxis('最高价',data_high,is_smooth=True,label_opts=opts.LabelOpts(is_show=False),markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="max")]))
price_line.set_global_opts(title_opts=opts.TitleOpts(title="茅台数据中的最高最低价的双折线图"),
                           yaxis_opts=opts.AxisOpts(type_="value",min_=500,max_=1250),
                           datazoom_opts = opts.DataZoomOpts()) # 由于时间范围跨度大，使用 DataZoomOpts 可以设置 X 轴的窗口，便于观察一段时间的数据
        # https://pyecharts.org/#/zh-cn/global_options?id=datazoomopts%ef%bc%9a%e5%8c%ba%e5%9f%9f%e7%bc%a9%e6%94%be%e9%85%8d%e7%bd%ae%e9%a1%b9
price_line.render_notebook()
```
